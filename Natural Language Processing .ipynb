```python
import pandas as pd 
messages=pd.read_csv('Downloads/smsspamcollection/SMSSpamCollection', sep='\t', names=['label', 'message'])
#/t will help in doing partition between tab separated values i.e. dependent and independent features 
```

```python
messages
```

```python
#data cleaning and data preprocessing 

corpus=[]
for i in range(0, len(messages)):
    review= re.sub('^[a-zA-Z]', ' ', messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
# creating bag of words 
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000) # only countvectorizer will have around 8000 columns but we want max features around 5000
X=cv.fit_transform(corpus).toarray()

y=pd.get_dummies(messages['label'])
```

```python
X.shape # 5000 most frequent words 
```

```python
y
```

```python
y=y.iloc[:,1].values   # we dont need two variables insted select one variable either 0 or 1
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
# Traning model using naive bayes classifer 
from sklearn.naive_bayes import MultinomialNB
spam_detect_model= MultinomialNB().fit(X_train, Y_train)
ypred=spam_detect_model.predict(X_test)
```

```python
ypred 
```

```python
from sklearn.metrics import confusion_matrix
confusion_m= confusion_matrix(Y_test,ypred)
```

```python
confusion_m
```

```python
from sklearn.metrics import accuracy_score 
accuracy=accuracy_score(Y_test, ypred)
```

```python
# preprocessing on paragraph 
text=re.sub(r'\[[0-9]*\]',' ',paragraph)
text=re.sub(r'\s+',' ', text)
text=text.lower()
text= re.sub(r'\d',' ', text)
text=re.sub(r'\s+',' ', text)
```

```python
sentences=nltk.sent_tokenize(text)
sentences=[nltk.word_tokenize(sentence) for sentence in sentences]
```
